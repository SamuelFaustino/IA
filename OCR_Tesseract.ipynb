```python
# Importar bibliotecas necessárias
import pytesseract
from PIL import Image
import requests
from io import BytesIO

# imagem de exemplo
# img = '/content/imagens/RÓTULO-lata.jpg'
import pytesseract
from PIL import Image
import os

# Diretório das imagens
diretorio = '/content/imagens/'

# Lista todos os arquivos no diretório
imagens = os.listdir(diretorio)

# Loop sobre cada imagem no diretório
for imagem_nome in imagens:
    # Caminho completo para a imagem
    imagem_caminho = os.path.join(diretorio, imagem_nome)

    # Verifica se o caminho é um arquivo e não um diretório
    if os.path.isfile(imagem_caminho):
        # Executar OCR na imagem
        texto = pytesseract.image_to_string(Image.open(imagem_caminho), lang='por')

        # Exibir o nome da imagem e o texto extraído
        print(f"Imagem: {imagem_nome}")
        print(texto)
        print("-" * 50)  # Apenas para separar a saída de diferentes imagens

# Executar OCR na imagem
# texto = pytesseract.image_to_string(img, lang='por')

# Exibir o texto extraído
# print(texto)
```
